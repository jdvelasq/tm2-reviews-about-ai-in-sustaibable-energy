```python
#
# Creates the id number for each node
id = 0
nodes_id = {}
for year in list(data.keys()):
    for cluster in range(len(data[year].keys())):
        key = str(year) + "." + str(cluster)
        nodes_id[key] = id
        id += 1
#
# Compute the link weights

links = {}

for year_curr in list(data.keys())[:-1]:
    #
    year_nxt = str(int(year_curr) + 1)
    for cluster_curr in range(len(data[year_curr].keys())):
        
        for word_curr in data[year_curr][str(cluster_curr)]:
            
            for cluster_nxt in range(len(data[year_nxt])):
                
                for word_nxt in data[year_nxt][str(cluster_nxt)]:

                    if word_curr == word_nxt:

                        key_curr = str(year_curr) + "." + str(cluster_curr)
                        key_nxt = str(year_nxt) + "." + str(cluster_nxt)
                        key = (key_curr, key_nxt)

                        if key not in links.keys():
                            links[key] = 0
                        links[key] += 1

#
# Creates the variables for the sankey diagram
source = [nodes_id[key[0]] for key in links.keys()]
target = [nodes_id[key[1]] for key in links.keys()]
value =  list(links.values())

label = [  
    "<br>".join([t.replace('_', ' ')  for t in data[year][str(cluster)][:5]])
    for year in list(data.keys()) 
    for cluster in range(len(data[year].keys()))
]



x = [  
    0.025 + 0.95 * (int(year) - 2011) / (2023-2011)
    for year in list(data.keys()) 
    for _ in range(len(data[year].keys()))
]

y = [
    #
    # 2011: 1
    0.2,
    # 2012: 3
    0.1, 0.22, 
    #
    # 2013: 1
    0.30, 
    #
    # 2014: 2
    0.21, 0.36, 
    #
    # 2015: 4
    0.09, 0.32, 0.55, 0.75, 
    #
    # 2016: 4
    0.03, 0.31, 0.52, 0.7, 
    #
    # 2017: 4
    0.11, 0.33, 0.57, 0.75, 
    #
    # 2018: 4
    0.01, 0.23, 0.47, 0.70, 
    #
    # 2019: 3
    0.09, 0.33, 0.55,
    #
    # 2020: 6
    0.09, 0.30, 0.53, 0.72, 0.83, 0.90, 
    #
    # 2021: 7
    0.05, 0.21, 0.34, 0.48, 0.63,  0.74, 0.95,
    #
    # 2022: 5
    0.08, 0.29, 0.47, 0.70, 1.02,
    #
    # 2023: 7
    0.08, 0.35, 0.54, 0.63, 0.80, 0.81,  0.9,
]

for year in list(data.keys()):
    for cluster in range(len(data[year].keys())):
        key = str(year) + "." + str(cluster)
        nodes_id[key] = id
        id += 1
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Sankey(
            arrangement = "fixed",
            node = dict(
                pad = 10,
                thickness = 10,
                line = dict(color = "black", width = 0.3),
                label = label,
                x=x,
                y=y,
            ),
            link = dict(
                source = source,
                target = target,
                value = value,
            ),
            textfont={"size":7, 'color': 'black'},
        )
    ]
)

for x_coor, year in enumerate(data.keys()):
    fig.add_annotation(
        x= 0.02 + 0.96 * x_coor / (len(data.keys()) - 1),
        y= 1.05,
        xref="paper",
        yref="paper",
        text=str(year),
        showarrow=False,
        font=dict(
            #family="Courier New, monospace",
            size=11,
            #color="black"
            ),
        align="center",
    )

fig.update_layout(width=1600, height=1000).show()
```
