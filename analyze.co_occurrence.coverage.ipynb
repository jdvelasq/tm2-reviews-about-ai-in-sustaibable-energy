```python
#
# article
#
for year in range(2011, 2023):
    print(f"---< {year} >" + "-" * 80)
    cov = coverage(
        #
        # PARAMS:
        field="author_keywords",
        #
        # DATABASE_PARAMS
        root_dir="./",
        database="main",
        year_filter=(year, year),
        cited_by_filter=(None, None),
    ).tail(15)
    print(cov)
    print()
```
